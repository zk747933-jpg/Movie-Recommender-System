```python
import ast
import pandas as pd

def safe_list(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return []

```

```python
for col in ['genres', 'keywords', 'cast', 'crew']:
    movies[col] = movies[col].apply(safe_list)
```

```python
def extract_names_safe(lst):
    # agar already names list hai → skip
    if lst and isinstance(lst[0], str):
        return lst
    return [i['name'] for i in lst]
```

```python
movies['genres']   = movies['genres'].apply(extract_names_safe)
movies['keywords'] = movies['keywords'].apply(extract_names_safe)
```

```python
movies['cast'] = movies['cast'].apply(
    lambda x: x if x and isinstance(x[0], str)
    else [i['name'] for i in x[:3]]
)

def extract_director_safe(lst):
    if lst and isinstance(lst[0], str):
        return lst
    for i in lst:
        if i.get('job') == 'Director':
            return [i['name']]
    return []

movies['crew'] = movies['crew'].apply(extract_director_safe)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps =  PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend('Avatar')
```
